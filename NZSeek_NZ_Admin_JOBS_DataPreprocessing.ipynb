```python
import pandas as pd
pd.set_option('display.width', 200) # print all the columns
#Display all the columns
pd.set_option('display.max_columns', None)
```

```python
# read data file
FileName = 'NZ_Admin_JOBS.xlsx'
Data = pd.read_excel(FileName,header=None)
# print(Data.head())
# Drop first row and reset index
Data = Data.drop(0,axis=0)
Data = Data.reset_index(drop=True)
Data
```

```python
# Data describe
Data.info()
# null values
Data.isnull().sum()
```

```python
# get rid of redumdent 'location:'
Data[3] = Data[3].map(lambda x: x.lstrip('location:'))
```

```python
# fill company name NaN in col 2 with Col 4
Data[2]=Data[2].fillna(Data[4].str.split(',', expand=True)[2])
Data
```

```python
# area
Data['area'] = Data[3].str.split(':',expand=True)[0]

# location
Data['Location'] =Data[3].str.split(':',expand=True)[1]

#drop col 3
Data = Data.drop([3],axis=1)
Data
```

```python
# Extract number of days from col 4
Data[4] = Data[4].str.split(',', expand=True)[0]
# # check data contain h replace to 1d
Data.loc[Data[4].str.contains('h'), 4] = '1d'

# extract number only from col 4
Data[4]=Data[4].str.extract('(\d+)')
Data
```

```python
# Extract classification & Hourly rate from column 5
Data = Data.assign(Classification=Data[5].str.extract('(classification.*)'))
Data = Data.assign(HourlyRate=Data[5].str.extract('(Competitive.*)'))
# Delete column 5
Data = Data.drop([5],axis=1)
Data
```

```python
# # Rename Data columns
Data.columns = ['Role','Website','Company','Reseased(Day)','area','Location','Classification','HourlyRate']
Data # 列中重复的值 无法去除
```
